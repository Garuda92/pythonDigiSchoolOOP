### 1. feladat:
+ #### Írj egy 'Teglalap' osztályt, ami tárolja a téglalap oldalainak hosszát (sz, m) és készíts hozzá metódusokat a terület , kerület számításhoz

+ #### Bővítsd egy 'negyzet' metódussal, ami logikai True értéket ad vissza ha négyzettel van dolgunk


```python
class Teglalap:
    def __init__(self,sz,m):
        self.sz = sz
        self.m = m
        
    def kerulet(self):
        #a teglalap terulete
        return 2*(self.sz + self.m)
    
    def terulet(self):
        #teglalap terulete
        return self.sz * self.m
    
    def IsNegyzet(self):
        #Megnézi a metódus, hogy a téglalap négyzet-e
        if (self.sz == self.m):
            return True
        else:
            return False
        
        

"""
innentől tesztelek
"""
t1 = Teglalap(5, 10)
t2 =Teglalap(4,4)

print(t1.terulet())
print(t1.kerulet())
print(t1.IsNegyzet())
print(t2.IsNegyzet())
        
```

### 2. feladat:
+ #### Írj egy 'Dolgozo' osztályt, ami tárolja a dolgozó nevét, fizetését és munkaidejét, (8 óra legyen alapértelmezésben) 
+ #### készíts hozzá __str__()  és __repr__() metódusokat 
+ #### készíts hozzá fizetés lekérdező és fizetést beállító metódusokat
+ #### készíts hozzá fizetés emelés metódust ami %-ban adja meg a béremelés mértékét


```python
"""
Írj egy 'Dolgozo' osztályt, ami tárolja a dolgozó nevét, fizetését és munkaidejét, (8 óra legyen alapértelmezésben)
készíts hozzá str() és repr() metódusokat
készíts hozzá fizetés lekérdező és fizetést beállító metódusokat¶
készíts hozzá fizetés emelés metódust ami %-ban adja meg a béremelés mértékét
"""

class Dolgozo():
    def __init__(self, nev,fizetes,munkaido = 8):
        self.nev = nev
        self.fizetes = fizetes
        self.munkaido = munkaido
        
    def __str__(self):
        #dolgozó objektum kiíratása
        return ' Egy dolgozó objektum vagyok, a nevem {self.nev} a fizetésem {self.fizetes} a munkaidőm napi {self.munkaido}'.format(self = self)
        
    def __repr__(self):
        return 'Dolgozo(nev={},fizetes={}, munkaido={})'.format(self.nev, self.fizetes, self.munkaido)
    """
    Getterek
    """    
    def getNev(self):
        return self.nev
   #fizetés lekérdezés 
    def getFizetes(self):
        return self.fizetes
    
    def getMunkaido(self):
        return self.munkaido
    """
    setterek
    """
    def setNev(self,nev):
        self.nev = nev
    #fizetés beállítás    
    def setFizetes(self,fizetes):
        self.fizetes = fizetes
        
        
    def setMunkaido(self,munkaido):
        self.munkaido = munkaido
        
        
    #fizetés emelés
    def fizetesEmeles(self,rat):
       """
       Százalékosan növeli a fizetését a dolgozónak
       Normál számmal kell megadni például, ha a bemeneti paraméter "10'
       az 10 % fizetésemelést jelent és a  függvény majd lebutitja a művelethez kellő formára        
       """
       rat = rat / 100
       self.fizetes = self.fizetes + self.fizetes * rat
       
#Tesztelés


d1 = Dolgozo('Beviz Elek', 8000000, 14)    
d2 = Dolgozo('Vic Elek', 200)

#meghívja az str függvényt, mégha nem is hívom rá az objektumra külön
print(d1)

print('-----------------------------------------------')

#repr esetén megkell hívni, illetve ha nincs str akkor ezt hívja meg
print(repr(d1))
print('-----------------------------------------------')
print(d1.getNev())
d1.setNev('Sanyikah')
print(d1.getNev())
print('-----------------------------------------------')
print(d1.getFizetes())
d1.fizetesEmeles(10)
print(d1.getFizetes())
```

### 3. feladat:

+ #### Készíts egy "Termosztat" osztályt, amivel termosztátok állapotát lehet kezelni.
+ #### A példány attributumai legyenek: "nev", "allapot" ("be" / "ki"), "homerseklet" 
+ #### Lehessen metódussal lekérdezni a hőmérsékletet 
+ #### Lehessen metódussal lekérdezni és beállítani az állapotot
+ #### készíts metódus amivel ellenkezőjére lehet bállitani az állapotot


```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jun 13 08:38:06 2020

@author: czeit
"""
"""
Készíts egy "Termosztat" osztályt, amivel termosztátok állapotát lehet kezelni.
A példány attributumai legyenek: "nev", "allapot" ("be" / "ki"), "homerseklet"
Lehessen metódussal lekérdezni a hőmérsékletet
Lehessen metódussal lekérdezni és beállítani az állapotot
készíts metódus amivel ellenkezőjére lehet bállitani az állapotot
"""

class Termosztat:
    def __init__(self,nev,allapot,homerseklet):
        self.nev = nev
        self.allapot = allapot
        self.homerseklet = homerseklet
        
    """
    Getterek
    """
    def getNev(self):
        return self.nev
    
    def getAllapot(self):
        return self.allapot
    
    def getHomerseklet(self):
        return self.homerseklet
    
    """
    Setterek
    """
    def setNev(self,nev):
        self.nev = nev
        
    def setAllapot(self,allapot):
        self.allapot = allapot
        
    def setHomerseklet(self,homerseklet):
        self.homerseklet = homerseklet
    """
    Funkcionalitás
    """    
    def switch(self):
        #termosztát kapcsolója csak Sheldon megne tudja
        if(self.allapot == 'ki'):
            self.allapot =='be'
        elif(self.allapot =='be'):
            self.allapot ='ki'
            
    
        
"""
Tesztelés
"""
t1 = Termosztat('Spyro', 'ki', 25)

print(t1.getAllapot())
t1.setAllapot('be')
print(t1.getAllapot())
t1.switch()
print(t1.getAllapot())
```
